```python
# 라벨 클래스 수정
import os
t_label_dir = "C:/Users/user/bee_dataset/set_origin/train/labels/" 
v_label_dir = "C:/Users/user/bee_dataset/set_origin/valid/labels/" 
t_label_list = os.listdir(t_label_dir)
v_label_list = os.listdir(v_label_dir)

def set_class_number(list, dir):
    for i in list:
        infos = []
        if i != "classes.txt":
            with open(dir + i, "r") as f:
                while True:
                    info = []
                    info = f.readline()
                    if not info:
                        break
                    info = info.split(" ")
                    class_num = int(info[0])
                    class_num -= 3
                    info[0] = str(class_num)
#                     print(i, info, class_num)
                    infos.append(info)
    
            
            with open(dir + i, "w") as f:
                for ins in infos:
                    cnt = 0
                    for i in ins:
                        info += (str(i))
                        cnt += 1
                        if cnt % 5 != 0:
                            info += " " 
                    
                f.write(info)
        
set_class_number(t_label_list, t_label_dir)
set_class_number(v_label_list, v_label_dir)
```

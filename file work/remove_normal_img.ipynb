#### 작성자 : 강성우

```python
import os
import shutil
```

```python
source_dir = "/home/kseongwoo/y2c/test/"
target_dir = "/home/kseongwoo/y2c/test_not_use/"
```

```python
testset = os.listdir(source_dir)
normal = []
n_count = 0
for data in testset:
    if 'txt' in data and data != "classes.txt": # txt 파일만 읽기
        with open(source_dir + data, 'r') as f:
            label = f.readline().split(" ")
            if len(label) < 2:
                print("노말: ", data)
                n_count += 1
                normal.append(data[:-4])

print(n_count)
```

```python
mv_count = 0
for n in normal:
    n_img = n + ".jpg"
    n_txt = n + ".txt"
    
    shutil.move(source_dir + n_img, target_dir + n_img)
    shutil.move(source_dir + n_txt, target_dir + n_txt)
    
    mv_count += 1
    print("이동 완료: ", n_img, ", ", n_txt)
    
print(mv_count)
```
